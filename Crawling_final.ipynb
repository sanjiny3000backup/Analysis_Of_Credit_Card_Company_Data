```python
# 카드 옵션 긁어오기
# 카드 사진 긁어오기
# 방법1
query_txt = new_card
txt = []
pic = []


driver = webdriver.Chrome('./chromedriver')
driver.get('https://www.shinhancard.com/conts/person/card_info/dream/credit/oil/1198245_46596.jsp')
time.sleep(2)

for q in query_txt:
    try:
        element = driver.find_element_by_class_name('tSerchLink')
        element.click()
        time.sleep(1)

        element = driver.find_element_by_xpath('//*[@id="query"]')
        element.send_keys(q)
        element.send_keys('\n')
        time.sleep(5)

        driver.find_element_by_xpath('//*[@id="loop_data02"]/li/div[2]/a').click()
        time.sleep(3)

        full_html = driver.page_source
        soup = BeautifulSoup(full_html, 'html5lib')
        content_list = soup.find_all("div", class_="shcd-card-detail-footer")
        txt.append(re.sub('\n', " ", content_list[0].get_text()))
        content_list = soup.find_all("dl", class_="shcd-card-brand")
        pic.append(content_list[0].select('img'))
        driver.back()
        time.sleep(1)
    except:
        txt.append(np.nan)
        pic.append(np.nan)
        driver.back()
```
