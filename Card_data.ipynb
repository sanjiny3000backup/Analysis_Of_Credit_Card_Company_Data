## crawling data load

```python
# crawling 데이터1
data_crawling=pd.read_csv("./Card_crawl.csv", index_col=0, encoding='cp949')
data_crawling["카드명"]=Card_data["카드명"]
data_crawling.head()
```

```python
# crawling 데이터2
data2_crawling=pd.read_csv("./Card_crawl2.csv", index_col=0, encoding='cp949')
data2_crawling.head()
```

```python
# crawl1 + crawl2
for tt in range(len(data_crawling)):
    for t in range(len(data2_crawling)):
        if data_crawling['카드명'][tt] == data2_crawling['카드명'][t]:
            data_crawling['options'][tt] = data2_crawling['options'][t]
            data_crawling['src'][tt] = data2_crawling['src'][t]
```

```python
# Card_data["Card_cp_crawl"]=data_crawling["src"]
# Card_data["Card_cp"]=np.where(pd.notnull(Card_data["Card_cp_crawl"])==True, Card_data["Card_cp_crawl"], Card_data["연회비"])
# Card_data.head()
```

```python
Card_data["옵션_crawl"]=data_crawling["options"]
Card_data["옵션_notNA"]=np.where(pd.notnull(Card_data["옵션"])==True, Card_data["옵션"], Card_data["옵션_crawl"])
Card_data.head()
```

```python
# 결측치 일부 크롤링으로 처리 후, 다시 결측치 확인
Card_data.isnull().sum()
```

> 결측치 처리 결과, 옵션_notNA 결측치 1개로 줄임

```python
# 최종 결측치 처리
Card_data["옵션_notNA"].fillna("옵션 없음", inplace=True)
```

## Option One-Hot-Encoding

```python
data_opt=Card_data["옵션_notNA"]
list_=[0 for k in range(len(data_opt))]

# 후불교통 옵션 포함시 +1
for j in range(len(data_opt)):
    if "교통" in str(data_opt[j]):
        list_[j]+=1

# 불가시 -1           
for t in range(len(data_opt)):
    if "후불교통 불가" in str(data_opt[t]):
        list_[t]-=1           

opt_cnt=pd.DataFrame({'후불교통':list_})
Card_data["후불교통"]=opt_cnt
```

```python
list_=[0 for k in range(len(data_opt))]
NO_family=["가족카드 발급 불가", "가족카드 불가", "가족카드불가"]

# 가족카드 옵션 포함시 +1
for j in range(len(data_opt)):
    if "가족" in str(data_opt[j]):
        list_[j]+=1

# 불가시 -1           
for t in range(len(data_opt)):
    for no in NO_family:
        if no in str(data_opt[t]):
            list_[t]-=1            

opt_cnt=pd.DataFrame({'가족카드':list_})
Card_data["가족카드"]=opt_cnt
```

```python
list_=[0 for k in range(len(data_opt))]

# 현금기능 옵션 포함시 +1
for j in range(len(data_opt)):
    if "현금" in str(data_opt[j]):
        list_[j]+=1

# 불가시 -1           
for t in range(len(data_opt)):
    if "현금기능 불가" in str(data_opt[t]):
        list_[t]-=1          

opt_cnt=pd.DataFrame({'현금기능':list_})
Card_data["현금기능"]=opt_cnt
```

```python
list_=[0 for k in range(len(data_opt))]

# IC기능 옵션 포함시 +1
for j in range(len(data_opt)):
    if "IC" in str(data_opt[j]):
        list_[j]+=1       

opt_cnt=pd.DataFrame({'IC기능':list_})
Card_data["IC기능"]=opt_cnt
```
